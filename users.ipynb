```python
import json
import ast
import re
import gzip
import pandas as pd
```

```python
row = []  # creo una lista vacia para ir guardando lo que recorre el bucle for

# creo un buble para recorrer el dataset
for i in gzip.open("users_items.json.gz"):
    # leo cada linea y la guardo en la lista row
    row.append(ast.literal_eval(i.decode("utf-8")))

# Creo el dataframe a partir de la lista
usuarios = pd.DataFrame(row)
usuarios
```

```python
usuarios = pd.json_normalize(row, record_path=['items'], meta=[
                             'steam_id', 'items_count', 'user_id', 'user_url'])
usuarios
```

```python
duplicados = usuarios.loc[usuarios.duplicated()]
duplicados
```

```python
usuarios_corregido = usuarios.drop_duplicates(keep='first')
usuarios_corregido
```

```python
guardar = 'user_limpio.csv'
usuarios_corregido.to_csv(guardar, index=False, encoding='utf-8')
```
